```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

save_path = r'/home/luan/Documentos/figuras_pesquisa/IC_Chemical_Potential/'
```

```python
Us = [5, 10]

fn = 'N6'
fn_prefixo = 'ED_'
fn_posfixo = 'U10.csv'

xs = []
mus = []
ChargeGap = []
for i in range(len(Us)):
    fn_posfixo = f'U{Us[i]}.csv'
    filenames = fn_prefixo + fn + fn_posfixo

    df = pd.read_csv(filenames, sep  = ',')

    x = np.array(df['x'])
    mu = np.array(df['mu'])
    CG = np.array(df['ChargeGap'])

    xs.append(x)
    mus.append(mu)
    ChargeGap.append(CG)
```

```python
with plt.rc_context({'xtick.color':'black', 'ytick.color':'black', 'figure.facecolor':'white'}):
    fig = plt.figure(figsize = (12, 8))
    for i in range(len(Us)):
        plt.scatter(xs[i], mus[i], label = r'$N_s$ = ' + f'{6}' + f', U = {Us[i]}', s = 30, alpha = 1-0.5*i)
        
    plt.plot(xs[i], xs[i], color = 'black', label = r'$f(x) = x$')
    plt.axhline(y = 0, linestyle = 'dashed', c = 'black', linewidth = 3, label = r'$\mu = 0$')

    plt.xlabel(r'$\dfrac{1}{2} - \dfrac{\epsilon_d}{U}$', fontsize = 25)
    plt.ylabel(r'$\dfrac{\mu}{U}$', fontsize = 25)
    plt.tick_params(labelsize = 20)

    plt.grid(visible = True)

    plt.legend(loc = 'upper left', fontsize = 20)

plt.savefig(save_path + 'ChemPotEDNs6.png', format = 'png', bbox_inches = 'tight')

plt.show()
```

```python
with plt.rc_context({'xtick.color':'black', 'ytick.color':'black', 'figure.facecolor':'white'}):
    fig = plt.figure(figsize = (12, 8))
    for i in range(len(Us)):
        plt.scatter(xs[i], ChargeGap[i], label = r'$N_s$ = ' + f'{6}' + f', U = {Us[i]}', s = 30, alpha = 1-0.5*i)

    plt.xlabel(r'$\dfrac{1}{2} - \dfrac{\epsilon_d}{U}$', fontsize = 25)
    plt.ylabel(r'$\dfrac{\mu}{U}$', fontsize = 25)
    plt.tick_params(labelsize = 20)

    plt.grid(visible = True)

    plt.legend(loc = 'lower center', fontsize = 20)

plt.savefig(save_path + 'ChargeGapEDNs6.png', format = 'png', bbox_inches = 'tight')

plt.show()
```
